```python
#@title Permanantly Delete a File or Folder from Google Drive

# @markdown Add extension for the file as like 'mp4', 'jpg' etc. If the file is an 'mp4', just write '.mp4' here. Leave it blank if it is a folder
Extension = '' #@param {type:"string"}

# @markdown Add the File or Folder names here
F1 = '1' #@param {type:"string"}
F2 = '' #@param {type:"string"}
F3 = '' #@param {type:"string"}
F4 = '' #@param {type:"string"}
F5 = '' #@param {type:"string"}
F6 = '' #@param {type:"string"}
F7 = '' #@param {type:"string"}
F8 = '' #@param {type:"string"}
F9 = '' #@param {type:"string"}
F10 = '' #@param {type:"string"}

filenames = [F1, F2, F3, F4, F5, F6, F7, F8, F9, F10]

from googleapiclient.discovery import build
from google.colab import auth

# Authenticate and create the drive service client
auth.authenticate_user()
drive_service = build('drive', 'v3')

for filename in filenames:
    if filename != '':
        # If no extension is provided, treat it as a folder
        if Extension == '':
            file_name = f'{filename}'
        else:
            file_name = f'{filename}{Extension}'

        # Search for the file or folder by name to retrieve its ID
        response = drive_service.files().list(q=f"name='{file_name}'").execute()

        # Check if the file or folder exists
        if 'files' in response and len(response['files']) > 0:
            file_id_to_delete = response['files'][0]['id']

            # Delete the file or folder by its ID
            drive_service.files().delete(fileId=file_id_to_delete).execute()
            print(f"'{file_name}' has been permanently deleted from Google Drive.")
        else:
            print(f"'{file_name}' not found in Google Drive.")
```

```python
# @title Permanantly Delete Everything from a Google Drive Folder

from googleapiclient.discovery import build
from google.colab import auth
import os
import re

def extract_folder_id_from_link(folder_link):
    # Extract the folder ID from a Google Drive folder link
    match = re.search(r'/folders/([^/?]+)', folder_link)
    if match:
        return match.group(1)
    else:
        return None

# Authenticate and create the drive service client
auth.authenticate_user()
drive_service = build('drive', 'v3')

# @markdown Add the Google Drive folder link
Link = 'https://drive.google.com/drive/folders/1Qjj95uW5TDlvY4ZcRuubsdMHYtVSlSy5?usp=drive_link' #@param {type:"string"}

# Extract the folder ID from the link
folder_id = extract_folder_id_from_link(Link)

# Get a list of all files in the folder
if folder_id:
    response = drive_service.files().list(q=f"'{folder_id}' in parents").execute()

    # Check if the folder contains any files
    if 'files' in response and len(response['files']) > 0:
        deleted_files = []
        for file in response['files']:
            # Delete the file by its ID
            drive_service.files().delete(fileId=file['id']).execute()
            deleted_files.append(file['name'])

        print(f"All files have been permanently deleted from the specified Google Drive folder: {', '.join(deleted_files)}")
    else:
        print("No files found in the specified folder.")
else:
    print("Failed to extract folder ID from the provided link.")
```
